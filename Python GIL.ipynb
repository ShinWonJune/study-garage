# Python GIL (Global Interpreter Lock)
- __정의__ : 여러 스레드가 파이썬 코드를 동시에 실행하지 못하도록 하는 규칙 (하나의 스레드 만이 bytecode를 실행할 수 있다)
    - Python은 thread safe 하지 않다. 여러 스레드가 하나의 자원에 접근하여 동시에 작업할 경우 어느 한쪽의 결과가 반영되지 않을 수 있다.(race conditioanl) 이를 방지하기 위한 장치
    

- 단, I/O 작업은 race conditinal 을 유발하지 않으므로 GIL 이 작용하지 않는다. 따라서 멀티스레딩이 좋은 효과를 낼 수 있다.
    - I/O : IN OUT, 외부 자원을 읽어들이는 작업

# Process and Thread
### Process 프로세스
- 정의 : 운영체제로부터 시스템 자원을 할당받는 작업의 단위(독립적인 개체)
    - 시스템 자원 : CPU시간, 주소 공간, 독립된메모리영역(code,data, stack, heap)
    

- 특징 : 각 프로세스는 서로의 변수나 자료구저에 접근할 수 없다.
    - 다른 프로세스의 자원에 접근하려면 프로세스 간의 통신(IPC, inter process communication) 을 사용해야 함

#### Multi-process 멀티프로세스
- 정의: 하나의 응용프로그램을 여러개의 프로세스로 구성하여, 각 프로세스가 하나의 작업(task)를 처리하도록 하는 것

- 장점: 여러 개의 자식 프로세스 중 하나에 문제가 발생하면 그 자식 프로세스만 죽는 것 이상의 영향이 확산되지 않는다.

- 단점: Context Switching 에서의 오버헤드
    - Context Switching 과정에서 캐쉬 메모리 초기화 등 무거운 작업이 진행되고 많은 시간이 소모되는 등의 오버헤드가 발생함
    - Context Switching: CPU에서 여러 프로세스를 돌아가면서 작업을 처리하는 과정
        - 동작중인 프로세스가 대기하면서 해당 프로세스의 상태(context)를 보관하고, 다음 순서의 프로세스가 동작하면 이전에 보관했던 프로세스의 상태를 복구하는 작업 (프로세스간 독립적인 메모리 때문에 변수, 데이터를 공유할 수 없다.)
        
        

### Thread 스레드 & 멀티스레드
- 정의: 프로세스 내의 실행 흐름 단위
- 특징: 
     1. 프로세스에 할당된 메모리 CPU등의 자원만 사용한다.
     2. Code,data,heap은 스레드간 공유하며 stack만 스레드마다 별도의 메모리를 할당한다. (따라서 한 스레드의 결과가 다른 스레드에 영향미친다.)
         - context switch시 stack영역만 처리하기 때문에 전환 속도가 빠르다
     3. 시스템 자원 소모 감소 및 처리량을 증가시킬 수 있다. (프로세스를 생성하는 것은 고비용, 여러 프로레스간 통신 부담 감소)
     4. 스레드는 기본적으로 __async 비동기__ 이다
        - 메인스레드에 .join() 하면 서브스레드가 종료될 때 까지 기다린다


- 주의: 멀티스레드는 디버깅이 어렵고, 자원공유문제(교착상태)가 발생할 수 있다.



#### Daemon thread 데몬스레드
- 정의 : 메인 스레드가 종료되면 즉시 종료되는 스레드

#### 호출한 함수외, 호출된 함수의 상호작용 방식에 따라 네가지로 구분 된다.
## 작업완료 시간 관리에 의한 분류
### 1. Syncronous 동기 (sync)
- 정의 : 호출한 함수가 호출된 함수의 작업 완료시간을 관리하는 경우 (호출된 함수의 작업 완료 시점까지 다른 작업을 중단하고 기다림)


### 2. Asyncronous 비동기 (async)
- 정의 : 호출된 함수의 작업 완료시간은 스스로가 관리하는 경우.
    - 호출된 함수는 callback을 받고, 작업이 완료되면 callback을 수행한다.(다른 객체의 인자로 넘어간다)
        - callback: 다른 코드의 인수로서 넘겨주는 실행 가능한 코드

## 제어권에 의한 분류
### 1. Block 
- 정의 : 호출된 함수가 작업이 끝날 때 까지 제어권을 가져, 호출한 함수는 호출된 함수의 작업이 끝날 때 가지 다른 작업을 수행하지 못하는 경우

### 2. Non-block
- 정의 : 호출 된 함수가 작업 완료 전에 제어권을 넘겨주어(`return`), 호출한 함수가 다른 작업을 수행할 수 있는 경우
