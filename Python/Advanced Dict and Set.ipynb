```python
# Dict Setdefault 예제

source = (('k1','val1'),   # Dict 으로 바꾸려고 함
          ('k1','val2'),   # key 값이 중복이 됨
          ('k2','val3'),   # 중복 key 끼리 묶어서 dict으로 표현 하고자 함
          ('k2','val4'),
          ('k2','val5'),
         )
new_dict1 = {}
new_dict2 = {}

#No use setdefault  : dict.append(value) 사용. 기존 value에 추가하기.
for k, v in source:
    if k in new_dict1:
        new_dict1[k].append(v)
    else:
        new_dict1[k] = [v] #list 형태로 담기

print(new_dict1)

# Use setdefault
for k, v in source:
    new_dict2.setdefault(k, []).append(v) # dict.setdefault(key, value) : 없는 key 값이면 key:value 등록 후 value 반환. 
    #있는 key값이면 해당 key값의 기존 값 return   

    print(new_dict2)

    
    
```

```python
# 사용자 정의 dict 상속 (UserDict 가능)


class UserDict(dict):             #dict를 받는 클래스를 생성하면, 대충 pair로 넣으면 dict라고 인식함.
    def __missing__(self,key):
        print('called : __missing__')
        if isinstance(key, str): # key가 있는데 __missing__ 메소드를 호출하면  KeyError를 발생ㅅ히킴 
            raise KeyError(key)  
        return self[str(key)]
    
    def get(self, key, default=None):
        print('called: __getitem__')
        try :
            return self[key]
        except KeyError:
            return default
    
    def __contains__(self, key):  #contains 함수는 in 으로 호출됨
        print('Called : __contains__')
        return key in self.keys() or str(key) in self.keys() #True or False로 반환

user_dict1 = UserDict(one = 1, two = 2)
user_dict2 = UserDict({'one ': 1, 'two':2})
user_dict3 = UserDict([('one',1),('two',2)])  #dict() 는 대충 pair container만 넣어주면 dictionary로 만들어준다
# 출력
# print
print(user_dict2.get('two'))
print('one' in user_dict3) 
```

```python
user_dict1.__missing__('one')  # key가 있는데 __missing__ 메소드를 호출하면  KeyError를 발생시킴 
```

```python
user_dict1['three'] #없는걸key를 key 접근하면 __missing__ 함수 호출됨
```

```python
print(user_dict1.get('three')) #get 함수를 통해 접근하면 None 반환함.
```
