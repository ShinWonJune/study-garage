### 프로그램 시간 재기

```python
%%time
import time
for _ in range(100):
    time.sleep(0.01)
#%%time 다음의 프로그램 실행 시간을 재는 듯
```

### 파이썬 파일 엑스포트하기

```python
%%writefile narutocode.py
def ninja(subject):
    for nin in subject:
        print(nin)
ninjas=['naruto','sikamaru','temari']

ninja(ninjas)
```
