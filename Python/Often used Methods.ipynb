```python
# __iter__

# iterable로 취급 할 떄 호출되는 method

# EX

class VectorP(object):
    def __init__(self,x,y):
        self.__x = float(x)  # 언더바 두개면, 변수를 숨기는 것. 직접 접근이 불가능
        self.__y = float(y)

    def __iter__(self):        # __iter__ 이터러블로 취급당할때 호출되는 함수!
        return (i for i in (self.__x, self.__y)) #Generator

    @property # property라는 데코레이터가 선행된다. v.x() 처럼 호출하지 않고 그냥  v.x 로 호출할 수 있다. => getter method
    def x(self):    # x로 접근하는 함수
        return self.__x

    @x.setter #setter method. gfetter로 정한 함수.setter  하는게 국룶
    def x(self, v):
        self.__x = float(v)
        
    @property
    def y(self):
        print('Called property y')
        return self.__y
    
    @y.setter
    def y(self, v):
        print('Calling setter y')
        if v < 30 :
            raise ValueError('Less than 30 is not available')
        self.__y = float(v)
# 객체 선언
v = VectorP(20,40)

for val in v:
    print(val)
```

```python
# __dict__

# 객체에 포함된 속성을 확인하는 method

# EX

print(v.__dict__) #클래스와, 변수 같이 붙여서 보여줌
```

```python
# __getitem__

# index로 속성에 접근가능하게 하는 메서드

#EX
class Objects:
    def __init__(self):
        self._numbers = [n for n in range(1,10000,3)]

    def __len__(self):
        return len(self._numbers)

    def __getitem__(self, idx): #index로 접근하는 method.
        return self._numbers[idx]

s = Objects()
s[100]
```

```python
# __mro__

print('EX5-3 -',Objects.__mro__) #클래스 구조를 순서대로 보여줌. 클래스 Objects는 클래스 object를 상속받는다.
```
