```python
# __iter__

# iterable로 취급 할 떄 호출되는 method

# EX

class VectorP(object):
    def __init__(self,x,y):
        self.__x = float(x)  # 언더바 두개면, 변수를 숨기는 것. 직접 접근이 불가능
        self.__y = float(y)

    def __iter__(self):        # __iter__ 이터러블로 취급당할때 호출되는 함수!
        return (i for i in (self.__x, self.__y)) #Generator

    @property # property라는 데코레이터가 선행된다. v.x() 처럼 호출하지 않고 그냥  v.x 로 호출할 수 있다. => getter method
    def x(self):    # x로 접근하는 함수
        return self.__x

    @x.setter #setter method. gfetter로 정한 함수.setter  하는게 국룶
    def x(self, v):
        self.__x = float(v)
        
    @property
    def y(self):
        print('Called property y')
        return self.__y
    
    @y.setter
    def y(self, v):
        print('Calling setter y')
        if v < 30 :
            raise ValueError('Less than 30 is not available')
        self.__y = float(v)
# 객체 선언
v = VectorP(20,40)

for val in v:
    print(val)
```

```python
# __getitem__

# index로 속성에 접근가능하게 하는 메서드

#EX
class Objects:
    def __init__(self):
        self._numbers = [n for n in range(1,10000,3)]

    def __len__(self):
        return len(self._numbers)

    def __getitem__(self, idx): #index로 접근하는 method.
        return self._numbers[idx]

s = Objects()
s[100]
```

```python
# __mro__

print('EX5-3 -',Objects.__mro__) #클래스 구조를 순서대로 보여줌. 클래스 Objects는 클래스 object를 상속받는다.
```

### `__code__` : 함수의 body를 나타내는 코드 객체.
- `dir(func.__code__)` : 함수 몸통의 dir을 보는 것
- `func.__code__.co_freevars` : Free variable 보여줌
- `__closure__` : 클로져의 참조값을 담고있는 변수. 

```python
def closure_avg1():
    # Free variable
    series = []

    #클로저 영역(body)
    def averager(v):
        series.append(v)        #클로저 영역밖의 변수를 계속 참조함. series 변수가 누적되어감. list.append 의 특성상. 참조만 했을 뿐인데 변수에 변화가 생김.
                                # (원래 클로져 영역 밖의 변수는 클로저 영역 안에서 재할당, 수정 불가능.)
        print('func >>> {}/{}'.format(series,len(series)))
        return sum(series) / len(series)
    return averager # () 없이 ,즉, 실행하지 않은 채로 함수 자체만 return 하는 것

avg_closure1 = closure_avg1()

print('EX5-1-', dir(avg_closure1))
print()
# __code__ : 컴파일된 함수의 body 를 나타내는 코드객체. 함수의 내용물. 즉. dir(func.__code__) 하면 함수 몸통의 attributes 보는것.
print('EX5-2-',dir(avg_closure1.__code__))  
print()
#free variable 호출
print('EX5-3-',avg_closure1.__code__.co_freevars) 
print()
# __closure__ : 클로져의 참조값을 담고있는 변수. 위 함수는 하나의 변수(series)를 참조하기때문에 길이 1인 튜플. __closure__[1] 없다는 뜻
print('EX5-4-1 ',avg_closure1.__closure__[0].cell_contents) 
```
