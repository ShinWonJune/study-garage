### CLASS , INSTANCE, STATIC METHODS
#### Class Method
- cls를 인자로 받는 method
- class를 통해서 바로 접근
    -   Student.raise_fee(arg)
#### Static Method
- self, cls 인자를 받지 않고 arg만 받는 method
- 인스턴스 속성, 인스턴스 메서드가 필요 없을 때 사용
- Class, instance 둘 다 접근 가능
    - Student.is_scholarship_st()
    - chulsoo.is_scholarship_st()

```python
class Student(object):
    '''
    student Class
    Author : Shin
    Data : 2021 11 22
    Description : Class, Static, Instance Method
    '''
    #Class Variable
    tuition_per = 1.0
    
    def __init__(self, id, first_name, last_name, email, grade, tuition, gpa):
        self._id = id
        self._first_name = first_name
        self._last_name = last_name
        self._email = email
        self._grade = grade
        self._tuition = tuition
        self._gpa = gpa
        
    #Instance method
    
    def full_name(self): #self 를 통해서 결과를 반환하는 것을 instance method 라 한다
        return '{} {}'.format(self._first_name, self._last_name)
    
    #Instance Method
    def detail_info(self):
        return 'Student Detail Info : {},{},{},{},{},{}'.format(self._id, self.full_name(), self._email,self._grade,self._tuition,self._gpa)
    
    #Instance Method
    def get_fee(self):
        return 'Before Tuition -> Id : {}, fee : {}'.format(self._id, self._tuition)
  
    # Instance Method
    def get_fee_culc(self):
        return 'After tuition -> Id :{}, fee : {}'.format(self._id, self._tuition*Student.tuition_per)

    def __str__(self):
        return 'Student Info -> name : {}, Grade : {} , Email : {}'.format(self._id, self._grade,self._email) 

    #class Method
    @classmethod #클래스를 받는 메쏘드 
    def raise_fee(cls, per):
        if per <= 1:
            print('Please 1 or more')
            return
        cls.tuition_per = per
        print('Succed! Tuition Increased!')
        
    #Class Method
    @classmethod #객체를 만드는 classmethod.. 첫 인자로 cls를 받는다.  This is the pythonic way!
    def student_const(cls, id, first_name, last_name, email, grade, tuition, gpa):
        return cls(id, first_name, last_name, email, grade, tuition*cls.tuition_per, gpa)
    
    #Static Method
    @staticmethod
    def is_scholarship_st(inst): #인스턴스에도 있고, 클래스에도 있는 함수 => instance.func  class.func 호출 가능
        if inst._gpa >= 4.3:
            return '{} is a scholarship recipient.'.format(inst.full_name())
        return 'Sorry. Not a scholarship recipient.'
```

### Class를 받아서 사용하는 함수는 @classmethod  라고 decorator 붙여주고 생성하면 됨

```python
# 장학금 혜택 여부(static method 미사용)
def is_scholarship(inst):
    if inst._gpa >= 4.3:
        return '{} is a scholarship recipient.'.format(inst.full_name())
    return 'Sorry. Not a scholarship recipient.'

print(is_scholarship(student_1))
print(is_scholarship(student_2))
print(is_scholarship(student_3))
print(is_scholarship(student_4))
```

```python
# Class를 통해 static method 사용
# 
print(Student.is_scholarship_st(student_1))
print(Student.is_scholarship_st(student_2))
print(Student.is_scholarship_st(student_3))
print(Student.is_scholarship_st(student_4))
```
