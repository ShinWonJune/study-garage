```python
import numpy as np
```

### reshape()

```python
# 1. x = np.reshape(a,(n, m))
# a를 nXm으로 reshape

a = np.array([1,10,2,3,4,5])
x = np.reshape(a,(2,3))
x
```

```python
# 2. x = a.reshape((n, m))
# a를 nXm으로 reshape

a = np.array([1,10,2,3,4,5])
x = a.reshape((2,3))
x
```

### index 호출

```python
# 1. np.where(a == 조건)
# array a 에서 조건에 해당하는 인자의 index를 araray tuple로 반환 --> [0] 붙여줘야 array만 추출

import numpy as np
a = np.array([[1,2,3],
             [1,2,3],
             [0,2,0]])

print(np.where(a[2,:] == a[0,:].argsort()[0:1][0]),'\n',np.where(a[2,:] == a[0,:].argsort()[0:1][0])[0])
```

```python
# 2. np.argmin(a)
# array a의 초소 요소의 index 반환
# 자매품 np.argmax 도 있음
np.argmin(a[0,:])
```

```python
# 3. np.argsort(a)
# 작은 순서대로 정렬시킨 후 index 반환

np.argsort(a[2,:])
```
